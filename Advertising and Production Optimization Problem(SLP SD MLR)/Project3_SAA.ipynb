```python
from __future__ import print_function
import sys
import cplex
import pandas as pd

f1 = pd.read_csv("data.csv")
f = ''.join(open("data.csv").readlines())
rhs1 = f.split(',')
sources = ['x1','x2']
second1 =[]
second2 =[]
```

```python
f1.error[0]
```

```python
for i in range(1,101):
    second1.append('ya'+str(i))
    second2.append('yb'+str(i))
len(second1)
```

```python
def ad():
    model = cplex.Cplex()
    model.objective.set_sense(model.objective.sense.minimize)
    model.variables.add(names = sources)
    model.variables.add(names = second1)
    model.variables.add(names = second2)
    model.variables.set_upper_bounds('x1',296.4)
    model.variables.set_upper_bounds('x2',49.6)
    model.variables.set_lower_bounds('x1',0.7)
    model.variables.set_lower_bounds('x2',0.0)
    model.linear_constraints.add(lin_expr=[cplex.SparsePair(ind=['x1','x2'],val=[1.0,1.0])], senses = ['L'],rhs = [200])
    model.linear_constraints.add(lin_expr=[cplex.SparsePair(ind=['x1','x2'],val=[-1.0,0.2])], senses = ['L'],rhs = [0])
    model.linear_constraints.add(lin_expr=[cplex.SparsePair(ind=['x1','x2'],val=[1.0,-0.7])], senses = ['L'],rhs = [0])
    
    for i in range(1,100):
        model.linear_constraints.add(lin_expr=[cplex.SparsePair(ind=[second1[i],second2[i]],val=[3.0,2.0])],senses=['L'],rhs = [36])
        model.linear_constraints.add(lin_expr=[cplex.SparsePair(ind=[second1[i],second2[i]],val=[1.0,0.0])],senses=['L'],rhs = [8])
        model.linear_constraints.add(lin_expr=[cplex.SparsePair(ind=[second1[i],second2[i]],val=[0.0,1.0])],senses=['L'],rhs = [12])
    for i in range(len(f1.error)):    
        model.linear_constraints.add(lin_expr=[cplex.SparsePair(ind=['x1','x2',second1[i],second2[i]],val=[-0.04585391,-0.18412361,1,1])],\
                                     senses=['L'],rhs = [3.146986251809624+f1.error[i]])
        
    model.objective.set_linear([('x1',0.5),('x2',0.2)])
    
    for i in range(1,100):
        model.objective.set_linear([(second1[i],-0.03),(second2[i],-0.05)])
    
    model.solve()
    print ()
    print ('Solution_status:' , model.solution.get_status())
    print ('Cost :' , model.solution.get_objective_value())
    print ('x1 :' , model.solution.get_values('x1'))
    print ('x2 :' , model.solution.get_values('x2'))
```

```python
ad()
```
